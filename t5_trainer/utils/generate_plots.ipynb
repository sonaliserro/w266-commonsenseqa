```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('seaborn-whitegrid')
```

### Transfer Learning Results: Accuracy of models over 10 epochs

```python
# x-axis as number of training steps
x = np.arange(start = 300, stop = 3100, step = 300)

# Accuracies over the training steps, taken from checkpoint_accuracies.md
t5_base = [0.566749,0.610156,0.610156,0.609337,0.628174,0.620803,0.631450,0.626536,0.624079,0.624898]
t5_cosmos = [0.552007,0.597052,0.609337,0.615889,0.624079,0.628174,0.624898,0.628993,0.630631,0.627355]
t5_social = [0.612613,0.615070,0.612613,0.609337,0.619984,0.624079,0.622441,0.615070,0.622441,0.622441]
t5_hellaswag = [0.564292,0.612613,0.606880,0.614251,0.616708,0.624079,0.621622,0.623260,0.623260,0.620803]
```

```python
plt.plot(x, t5_base, color='blue', marker='.', label='T5-base')
plt.plot(x, t5_cosmos, color='green', linestyle='dashed', marker='.', label='T5-base-CosmosQA')
plt.plot(x, t5_social, color='red', linestyle='dashed', marker='.', label='T5-base-SocialIQA')
plt.plot(x, t5_hellaswag, color='orange', linestyle='dashed', marker='.', label='T5-base-HellaSwag')
plt.title('Model Performance over 10 epochs')
plt.xlabel('Training Steps')
plt.ylabel('Accuracy')
plt.legend();
```
